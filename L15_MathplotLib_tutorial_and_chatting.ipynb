```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[2,4,5]
y=[23,33,44]
x2=[3,5,7]
y2=[3,5,6]

plt.plot(x,y,label='FirstLine',linewidth=4)
plt.plot(x2,y2,label='SecondLine' ,linewidth=3)
plt.legend()
plt.show()
```
